```python
import os
import datetime
import random

print("---- Intermediate Python Core Exercise ----")
print("This exercise covers Python fundamentals, data structures, functions,")
print("core libraries (os, datetime, random), file handling, and exception handling.")
print("Estimated completion time: 3-4 hours.")
print("-------------------------------------------")

# ----- Global Data for the Exercise -----

DEVICE_TYPES = ['Temperature_Sensor', 'Humidity_Sensor',
                'Pressure_Sensor', 'Light_Sensor']

LOCATIONS = ['Lab1', 'Lab2', 'OfficeA', 'OfficeB', 'Warehouse']

READING_RANGES = {
    'Temperature_Sensor': (18.0, 30.0),
    'Humidity_Sensor': (30.0, 80.0),
    'Pressure_Sensor': (980.0, 1030.0),
    'Light_Sensor': (100, 1000)
}

print("\n--- Task 1: Data Generation and Basic Structures ---")

# =========================
# Task 1.1
# =========================

raw_sensor_data = []

base_time = datetime.datetime.now()

for i in range(100):
    timestamp = (base_time + datetime.timedelta(minutes=i)).strftime("%Y-%m-%d %H:%M:%S")
    device_id = f"DEV{i+1:03d}"
    device_type = random.choice(DEVICE_TYPES)
    location = random.choice(LOCATIONS)

    low, high = READING_RANGES[device_type]
    value = round(random.uniform(low, high), 2)

    reading = (timestamp, device_id, device_type, location, value)
    raw_sensor_data.append(reading)

print("Generated", len(raw_sensor_data), "sensor readings.")

# =========================
# Task 1.2
# =========================
print("\ntask-2")
temperature_readings = []

for reading in raw_sensor_data:
    if reading[2] == "Temperature_Sensor":
        temp = {
            "timestamp": reading[0],
            "device_id": reading[1],
            "location": reading[3],
            "temperature_celsius": reading[4]
        }
        temperature_readings.append(temp)

print("Temperature readings:", len(temperature_readings))

# =========================
# Task 1.3
# =========================
print("\ntask-3")
unique_locations = set()

for reading in raw_sensor_data:
    unique_locations.add(reading[3])

print("Unique Locations:")
print(unique_locations)

# =========================
# Task 1.4
# =========================
print("\ntask-4")
device_stats = {}

for device in DEVICE_TYPES:
    device_stats[device] = (float('inf'), float('-inf'))

for reading in raw_sensor_data:
    device = reading[2]
    value = reading[4]

    min_val, max_val = device_stats[device]

    if value < min_val:
        min_val = value

    if value > max_val:
        max_val = value

    device_stats[device] = (min_val, max_val)

print("Device Statistics")
for device, stats in device_stats.items():
    print(device, ":", stats)

# =========================
# Task 1.5
# =========================
print("\ntask-5")
print("--- Task 1.5: Advanced Indexing and Slicing ---")

sample_string = "PythonProgrammingIsFun"

# ---------- Task 1.5.1 ----------

sliced_readings_1 = raw_sensor_data[9:21]
sliced_readings_2 = raw_sensor_data[::5]

print("\nTask 1.5.1")
print("10th to 20th readings:", len(sliced_readings_1))
print("Every 5th reading:", len(sliced_readings_2))

# ---------- Task 1.5.2 ----------

substring_1 = sample_string[6:17]
substring_2 = sample_string[5::-1]

print("\nTask 1.5.2")
print("Substring 1:", substring_1)
print("Substring 2:", substring_2)

# ---------- Task 1.5.3 ----------

fifth_reading_device_type = raw_sensor_data[4][2]
device_type_suffix = fifth_reading_device_type[-6:]

print("\nTask 1.5.3")
print("5th Reading Device Type:", fifth_reading_device_type)
print("Suffix:", device_type_suffix)
# Task 2: Functions
print("\n--- Task 2: Function Definitions and Usage ---")

print("Task 2.1")
def get_readings_by_location(readings_list, target_location):
    result = []
    for reading in readings_list:
        if reading[3] == target_location:
            result.append(reading)
    return result

lab1_readings = get_readings_by_location(raw_sensor_data, "Lab1")
print("Lab1 Readings:", len(lab1_readings))


print("Task 2.2")
def calculate_average_temperature(temp_data_list):
    if len(temp_data_list) == 0:
        return 0.0

    total = 0
    for item in temp_data_list:
        total += item["temperature_celsius"]

    return round(total / len(temp_data_list), 2)

avg_temp = calculate_average_temperature(temperature_readings)
print("Average Temperature:", avg_temp)


print("Task 2.3")
def generate_report_line(reading_dict):
    return f"[{reading_dict['timestamp']}] Device {reading_dict['device_id']} at {reading_dict['location']}: {reading_dict['temperature_celsius']}°C"

if temperature_readings:
    print(generate_report_line(temperature_readings[0]))
#task 3: Core Libraries
print("\n--- Task 3: Core Library Usage ---")

print("Task 3.1")
reports_dir = "sensor_reports"

if not os.path.exists(reports_dir):
    os.makedirs(reports_dir)
    print("Directory created.")
else:
    print("Directory already exists.")


print("Task 3.2")
if len(raw_sensor_data) > 1:

    first_time = datetime.datetime.strptime(
        raw_sensor_data[0][0],
        "%Y-%m-%d %H:%M:%S"
    )

    last_time = datetime.datetime.strptime(
        raw_sensor_data[-1][0],
        "%Y-%m-%d %H:%M:%S"
    )

    duration = last_time - first_time

    days = duration.days
    hours = duration.seconds // 3600
    minutes = (duration.seconds % 3600) // 60

    print("Duration:", days, "days", hours, "hours", minutes, "minutes")


print("Task 3.3")
random_values = []

for i in range(5):
    random_values.append(random.randint(100, 200))

print("Random Values:", random_values)
# Task 4: File Handling
print("\n--- Task 4: File I/O ---")

print("Task 4.1")
report_filename = os.path.join(reports_dir, "temperature_report.txt")

try:
    with open(report_filename, "w") as f:
        for reading in temperature_readings:
            f.write(generate_report_line(reading) + "\n")

    print("Temperature report written successfully.")

except IOError as e:
    print("Error:", e)


print("Task 4.2")
try:
    with open(report_filename, "r") as f:
        read_report_lines = f.readlines()

    print("\nFirst 3 Lines:")
    for line in read_report_lines[:3]:
        print(line.strip())

except IOError as e:
    print("Error:", e)


print("\n--- Exercise Complete! ---")

```
